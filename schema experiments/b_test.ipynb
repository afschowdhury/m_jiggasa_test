```python
import matplotlib.pyplot as plt

# Set the font family and size
# plt.rcParams['font.family'] = 'Noto Sans Bengali UI'
plt.rcParams['font.family'] = 'Noto Sans Bengali'

plt.rcParams['font.size'] = 16

# Create a plot
fig, ax = plt.subplots()
ax.plot([1, 2, 3])

# Set the x-axis label to Bengali text
ax.set_xlabel('বাংলা টেক্সট')
ax.set_ylabel('English Text')

plt.show()
```

```python
import matplotlib.font_manager as fm

# Find the font file
font_path = 'aums.ttf'

# Add the font to the font manager
fm.FontManager.addfont( path =font_path)

```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(range(10))

prop = fm.FontProperties(fname='Nikosh.ttf')
ax.set_xlabel('বাংলা লিখা ব্যাঙ্গ', fontproperties=prop, size=32)
ax.set_ylabel('Test', fontproperties=prop, size=32)


plt.show()
```
