## Loading QA model

```python
qa_model_location = "/content/gdrive/MyDrive/DL_models/language_model_bn/xlm-roberta-large-squad2-test-1/afs_xlm_roberta"
```

```python
from transformers import  pipeline

qa_model = pipeline("question-answering",model = qa_model_location,tokenizer=qa_model_location)
```

```python
import torch
from sentence_transformers.util import semantic_search

def context_query(query):
  q_emb = retriver_model.encode(query)

  query_embeddings = torch.FloatTensor(q_emb)
  hits = semantic_search(query_embeddings, context_embeddings, top_k = 7)
  print(f"User's Question: {query}")
  print("Most Similar Context Found: ")
  for i in range(len(hits[0])):
    print(f"Context {i}:" ,contexts[hits[0][i]["corpus_id"]] )
    # for item in tests:
    #   if item['question'] == questions[hits[0][i]["corpus_id"]]:
    context = contexts[hits[0][i]["corpus_id"]]
    qa_input = {'question': query,'context': context}
    res = qa_model(qa_input)
    print(res)
```

```python
import torch
from sentence_transformers.util import semantic_search

def context_query_only_ans(query):
  q_emb = retriver_model.encode(query)
  respns = []

  query_embeddings = torch.FloatTensor(q_emb)
  hits = semantic_search(query_embeddings, context_embeddings, top_k = 7)
  print(f"User's Question: {query}")
  # print("Most Similar Context Found: ")
  for i in range(len(hits[0])):
    # print(f"Context {i}:" ,contexts[hits[0][i]["corpus_id"]] )
    # for item in tests:
    #   if item['question'] == questions[hits[0][i]["corpus_id"]]:
    context = contexts[hits[0][i]["corpus_id"]]
    qa_input = {'question': query,'context': context}
    
    res = qa_model(qa_input)
    respns.append(res)
    # print(res)
  scores = []
  for ans in respns:
    scores.append(ans['score'])
  max_score = max(scores)
  for res in respns:
    if max_score == res['score']:
      print(f"{'*'*15} QA models best answer:  {'*'*15}")
      print(res['answer'])
```

```python
import random

def query_select():
  idx = random.randint(0,len(q1s)-1)

  if type(q1s[idx])!= str:
    return query_select()
  else:
    return q1s[idx] , idx
```

```python
def generate_test():

  query,idx = query_select()
  context_query(query)
  print("----------Actual answer :-------------------")
  print(f"\t {e_a[idx]}")
  
```

```python
def generate_test_1s():

  query, idx = query_select()
  context_query_only_ans(query)
  print(f"{'-'*15} Actual answer {'-'*15}")
  print(e_a[idx])
```
