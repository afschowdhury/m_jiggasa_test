```python
import matplotlib.font_manager as fm


# fig, ax = plt.subplots()
# ax.plot(range(10))

# prop = fm.FontProperties(fname='Nikosh.ttf')
# ax.set_xlabel('বাংলা লিখা ব্যাঙ্গ', fontproperties=prop, size=32)
# ax.set_ylabel('Test', fontproperties=prop, size=32)


# plt.show()


# Set the font family and size
# plt.rcParams['font.family'] = "Arial Unicode MS"
# plt.rcParams['font.size'] = 16

def plot_bar_plot(x,y,xlabel,ylabel,title,show_number=False,show_x=False,rotation='vertical',t_f_s = 20,figsize=(25,6)):
    
    
    
    fig , ax = plt.subplots(figsize=figsize)
    # y = [total_unique_ques,total_mujib_ques,total_no_mujib_ques]
    x = x
    y = y
    
    colors = plt.cm.Dark2(range(6))
    
    prop = fm.FontProperties(fname='Nikosh.ttf')

    bar_plot = ax.bar(x,y,color = colors)

    ax.set(xlabel= xlabel,
        ylabel = ylabel);
    
    if show_x:
        ax.set_xticks(ticks = x, fontproperties=prop,labels= x, rotation=rotation)#,fontproperties=font) 
    
    if show_number:
        ax.bar_label(bar_plot,padding=3)
    fig.suptitle(t=title,
             size = t_f_s)
```

```python
plot_bar_plot(list(ques_number.keys()),
              list(ques_number.values()),
              "Q's",
              "# of Questions",
              "# of Questions in Q's",
              show_number=True)
```
