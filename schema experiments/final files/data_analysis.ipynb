## Defining Plotting Function

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.style.available

plt.style.use(plt.style.available[4])
```

```python
def plot_bar_plot(x,y,xlabel,ylabel,title,show_number=False,show_x=False,rotation='vertical',t_f_s = 20,figsize=(25,6)):
    
    
    
    fig , ax = plt.subplots(figsize=figsize)
    
    x = x
    y = y
    
    colors = plt.cm.Dark2(range(6))
    
    # prop = fm.FontProperties(fname='Nikosh.ttf')
    prop = fm.FontProperties(fname='data/fonts/Nikosh.ttf')


    bar_plot = ax.bar(x,y,color = colors)

    ax.set(xlabel= xlabel,
        ylabel = ylabel);
    
    if show_x:
        ax.set_xticks(ticks = x, fontproperties=prop,labels= x, rotation=rotation)#,fontproperties=font) 
    
    if show_number:
        ax.bar_label(bar_plot,padding=3)
    fig.suptitle(t=title,
             size = t_f_s)
```

## Question By Categories


```python
ques_df = df[['Q1','Q2','Q3','Q4','Mujib Relevant']]
ques_df.head()
```

```python
mujib_relevant_ques = ques_df[ques_df['Mujib Relevant'] == "YES"]
mujib_relevant_ques
```

```python
total_mujib_ques = len(mujib_relevant_ques)
total_unique_ques = len(ques_df)
total_no_mujib_ques = total_unique_ques- total_mujib_ques

```

```python
plot_bar_plot(["# Unique Questions","# Mujib Relevant Question","# Not relevant To Mujib Questions"],
              [total_unique_ques,total_mujib_ques,total_no_mujib_ques],
              "Questions Category",
              "# of Questions",
              "Number of Questions ( Mujib relevant Vs Non-relevant)",
              show_number=True
              )
```

## Number of Questions vs Context

```python
ctx_ques = []

for item in data:
    ctx = item['context']
    qtns = []
    for qas in item['qas']:
        qtns.append(qas['qtns'])
    ctx_ques.append({
        'ctx':ctx,
        'qtns': qtns
        
    })
```

```python
qs_in_ctx = []
for item in ctx_ques:
    qs_in_ctx.append(len(item['qtns']))
```

```python
len(ctx_ques) , len(qs_in_ctx)
```

```python
plot_bar_plot([*range(1,len(ctx_ques)+1,1)],
              qs_in_ctx,
              "Context #",
              "# of Questions",
              "Question Per Context",
              show_number=True,
              show_x=True)
```
