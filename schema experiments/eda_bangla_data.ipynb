```python


def plot_bar_plot(x,y,xlabel,ylabel,title,show_number=False,show_x=False,rotation='vertical',t_f_s = 20,figsize=(25,6)):
    fig , ax = plt.subplots(figsize=figsize)
    # y = [total_unique_ques,total_mujib_ques,total_no_mujib_ques]
    x = x
    y = y

    bar_plot = ax.bar(x,y)

    ax.set(xlabel= xlabel,
        ylabel = ylabel);
    
    if show_x:
        ax.set_xticks(ticks = x, labels= x, rotation=rotation) 
    
    if show_number:
        ax.bar_label(bar_plot,padding=3)
    fig.suptitle(t=title,
             size = t_f_s)
```

```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

# fig, ax = plt.subplots()
# ax.plot(range(10))

# prop = fm.FontProperties(fname='Nikosh.ttf')
# ax.set_xlabel('বাংলা লিখা ব্যাঙ্গ', fontproperties=prop, size=32)
# ax.set_ylabel('Test', fontproperties=prop, size=32)


plt.show()


# Set the font family and size
# plt.rcParams['font.family'] = "Arial Unicode MS"
# plt.rcParams['font.size'] = 16

def plot_bar_plot(x,y,xlabel,ylabel,title,show_number=False,show_x=False,rotation='vertical',t_f_s = 20,figsize=(25,6)):
    fig , ax = plt.subplots(figsize=figsize)
    # y = [total_unique_ques,total_mujib_ques,total_no_mujib_ques]
    x = x
    y = y
    
    colors = plt.cm.Dark2(range(6))
    
    prop = fm.FontProperties(fname='Nikosh.ttf')

    bar_plot = ax.bar(x,y,color = colors)

    ax.set(xlabel= xlabel,
        ylabel = ylabel);
    
    if show_x:
        ax.set_xticks(ticks = x, fontproperties=prop,labels= x, rotation=rotation)#,fontproperties=font) 
    
    if show_number:
        ax.bar_label(bar_plot,padding=3)
    fig.suptitle(t=title,
             size = t_f_s)
```

```python
from collections import Counter
from tqdm.notebook import tqdm_notebook
word_ls = []

for sen in contexts.tolist():
    word_ls.extend(sen.split())
    # print(len(word_ls))
     

print(len(word_ls))
     

Counter = Counter(word_ls)
most_occured_20 = Counter.most_common(20)
# print(most_occured_20)
     

most_occured_50 = Counter.most_common(50)
print(most_occured_50)
```
